### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
print("\nTienda ")
print(tienda.columns)

print("\nTienda ")
print(tienda2.columns)
print("\nTienda ")
print(tienda3.columns)
print("\nTienda ")
print(tienda4.columns)
```

```python
#ganancia_1 = sum(tienda['Precio']

def calculo_ganancias(tiendaselec):
  ingreso = sum(tiendaselec['Precio'])
  gastos= sum(tiendaselec['Costo de envío'])

  ganancia = ingreso-gastos
  return ganancia

"""
"""
ganancia1 = int(calculo_ganancias(tienda))
print(f"La ganancia para la tienda 1 es de: ${ganancia1:,}")

ganancia2= calculo_ganancias(tienda2)
print(f"La ganancia para la tienda 2 es de: ${ganancia2:,}")

ganancia3= calculo_ganancias(tienda3)
print(f"La ganancia para la tienda 3 es de: ${ganancia3:,}")

ganancia4= calculo_ganancias(tienda4)
print(f"La ganancia para la tienda 4 es de: ${ganancia4:,}")



```

```python
import matplotlib.pyplot as plt


tiendas= [ "Tienda 1","Tienda 2","Tienda 3","Tienda 4"]
ganancias= [ganancia1,ganancia2,ganancia3,ganancia4]

plt.bar(x=tiendas, height=ganancias,color="#951855")
plt.title("Ganancias por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ganancia ($)")
plt.grid(axis='y', linestyle='--', alpha=0.8)
plt.show()
```

```python
indice_min = ganancias.index(min(ganancias))
print(f'La tienda {indice_min + 1} genera menos ganancias con un total de ${min(ganancias)}')

```

# 3. Calificación promedio de la tienda

# 4. Productos más y menos vendidos

```python
def productos (tiendaSelec):
  ventas_por_producto = tiendaSelec['Producto'].value_counts()
  mas_vendido = ventas_por_producto.idxmax()
  cantidad_max = ventas_por_producto.max()
  menos_vendido = ventas_por_producto.idxmin()
  cantidad_min = ventas_por_producto.min()
  return mas_vendido,cantidad_max, menos_vendido, cantidad_min

lista_tienda=productos(tienda)
lista_tienda2=productos(tienda2)
lista_tienda3=productos(tienda3)
lista_tienda4=productos(tienda4)
lista_tiendas= [lista_tienda, lista_tienda2, lista_tienda3,lista_tienda4]

count=1
listaFinalProductos=[]
for i in lista_tiendas:
  print("")
  print(f"Tienda {count}")
  print(f"Producto más vendido: {i[0]} ({i[1]} ventas)")
  print(f"Producto menos vendido: {i[2]} ({i[3]} ventas)")
  count+=1

```

# 5. Envío promedio por tienda

```python
def prom_envio(tiendaSelec):
  promedio= sum(tiendaSelec['Costo de envío'])/len(tiendaSelec['Costo de envío'])
  return promedio

print("PROMEDIO DE GASTOS DE ENVÍO POR TIENDA")
print("")
promedio1 = int(prom_envio(tienda))
print(f"El promedio de gastos de envío de la tienda 1 es de : ${promedio1:,}")
print("")
promedio2 = int(prom_envio(tienda2))
print(f"El promedio de gastos de envío de la tienda 2 es de : ${promedio2:,}")
print("")
promedio3 = int(prom_envio(tienda3))
print(f"El promedio de gastos de envío de la tienda 3 es de : ${promedio3:,}")
print("")
promedio4 = int(prom_envio(tienda4))
print(f"El promedio de gastos de envío de la tienda 4 es de : ${promedio4:,}")



```

```python
#Grafica promedio envio
promedios=[promedio1,promedio2,promedio3,promedio4]
colores=["#b4aaef","#9cedfb","#d3ffce","#ffc1c1"]

plt.pie(promedios, labels=tiendas, autopct='%1.1f%%', colors=colores)
plt.title("Promedio de calificaciones por tienda")
plt.show()
```
